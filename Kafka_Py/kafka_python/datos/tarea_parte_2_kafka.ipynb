Ejercicio 2. Envío estructurado a Kafka
Crear un topic empresas y enviar cada fila del CSV como mensaje. El formato del mensaje debe permitir reconstruirse después sin problemas.

Lectura fichero

```python
from confluent_kafka import Producer
import pandas as pd

csv_file_path = 'data_empresa.csv'
df = pd.read_csv(csv_file_path)

producer = Producer({'bootstrap.servers': 'localhost:9092'})

for _, row in df.iterrows():
    message = f"{row['empresa']},{row['pais']},{row['precioLY]},{row['precioCY]}"
    producer.produce('testtopic', value=message)

producer.flush()
```
